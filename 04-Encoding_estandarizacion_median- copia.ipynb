```python
import pandas as pd

from sklearn.preprocessing import StandardScaler

import warnings
warnings.filterwarnings('ignore')
```

```python
df= pd.read_csv("datos/04-bikes_mediana.csv", index_col= 0).reset_index(drop= True)
df.head()
```

### Probamos primero estas opciones
- **Year** : El encoding estaría bien hecho ya porque el año 1 hay más bicis alquiladas
- **Workingday** : El encoding también estaría bien
- **Weathersit** : El encoding sería 1 - 2, para el 2 le ponemos un 1 y para el 3 le ponemos un 0
- **Season** : Hay que hacerlo. Verano 2, primavera y otoño 1, winter 0
- **mnth** enero: 0, febrero, marzo, abril, noviembre y diciembre: 1, junio, julio, agosto, septiembre, octubre: 2
- **Weekday** : Hacemos un map diferente para registered y para casual

```python
map_holiday = {0:7, 1:0}
df["holiday"] = df["holiday"].map(map_holiday)
```

```python
map_weathersit = {3:0, 2:2, 1:3}
df["weathersit"] = df["weathersit"].map(map_weathersit)
```

```python
map_season = {"winter":0, "autumn":2, "spring":2, "summer":3}
df["season"] = df["season"].map(map_season)
```

```python
mapa_weekday = {1: 0, 2:0, 3: 1, 4: 2, 5: 2, 6:2, 7:0}
df["weekday"] = df["weekday"].map(mapa_weekday)
```

```python
mapa_mnth = {1: 0, 2:1, 3: 1, 4: 1, 5: 2, 6:3, 7:3, 8:4, 9:4, 10:3, 11:1, 12:1}
df["mnth"] = df["mnth"].map(mapa_mnth)
```

```python
map_workingday = {0:1, 1:2}
df["workingday"] = df["workingday"].map(map_workingday)
```

```python
df.head()
```

```python
df.to_csv("datos/08-bikes_encoding_sinest_mediana.csv")
```

```python
df_est = df.copy()
df_est.head(1)
```

## ESTANDARIZACION

```python
num = df_est[["atemp", "hum", "windspeed"]]
num.head(2)
```

```python
st = StandardScaler()
x_tranformadas = pd.DataFrame(st.fit_transform(num))
x_tranformadas.columns = num.columns
df_est[x_tranformadas.columns] = x_tranformadas
df_est.head()
```
